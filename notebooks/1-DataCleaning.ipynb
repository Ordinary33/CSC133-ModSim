### Columns to keep
1. requested_processors (SWF Col 8)
2. requested_time (SWF Col 9)
3. requested_memory (SWF Col 10 - Imputed)
4. queue_number (SWF Col 15 - Encoded)
5. preceding_job_number (SWF Col 17 - Imputed)
6. think_time (SWF Col 18 - Imputed)
7. current_cpu_utilization (Engineered)

```python
col_to_keep = [
    'requested_processors', 'requested_time', 'requested_memory', 
    'queue_number', 'preceding_job_number', 'think_time',
    
    'wait_time',
    
    'submit_time', 'run_time', 'allocated_processors',
    
    'user_id' 
]
```

```python
df_working = df_clean[col_to_keep]
```

```python
df_working = df_working[df_working['wait_time'] >= 0]
```
