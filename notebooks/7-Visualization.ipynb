```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../result/final_monte_carlo_results.csv')
```

```python
fig, axes = plt.subplots(3, 1, figsize=(8, 12), sharex=True)

# Custom colors and line styles
baseline_color = '#d62728' # Solid Red
xgb_color = '#1f77b4'      # Solid Blue

# ==========================================
# SUBPLOT 1: Mean Wait Time
# ==========================================
axes[0].plot(df['Iteration'], df['FCFS_Wait_Hrs'], color=baseline_color, label='Standard EASY', linewidth=2, marker='.', alpha=0.8)
axes[0].plot(df['Iteration'], df['XGB_Wait_Hrs'], color=xgb_color, label='XGBoost-Guided', linewidth=2, marker='.')
axes[0].set_title('Mean Wait Time per Workload', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Hours', fontsize=12)
axes[0].grid(True, linestyle='--', alpha=0.6)
axes[0].legend(loc='upper right')

# ==========================================
# SUBPLOT 2: Bounded Slowdown (BSLD)
# ==========================================
axes[1].plot(df['Iteration'], df['FCFS_BSLD'], color=baseline_color, linewidth=2, marker='.', alpha=0.8)
axes[1].plot(df['Iteration'], df['XGB_BSLD'], color=xgb_color, linewidth=2, marker='.')
axes[1].set_title('Bounded Slowdown (BSLD)', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Ratio (Lower is Better)', fontsize=12)
axes[1].grid(True, linestyle='--', alpha=0.6)

# ==========================================
# SUBPLOT 3: CPU Utilization
# ==========================================
axes[2].plot(df['Iteration'], df['FCFS_Util_%'], color=baseline_color, linewidth=2, marker='.', alpha=0.8)
axes[2].plot(df['Iteration'], df['XGB_Util_%'], color=xgb_color, linewidth=2, marker='.')
axes[2].set_title('Average CPU Utilization', fontsize=14, fontweight='bold')
axes[2].set_xlabel('Simulation Iteration (Workload 1 - 50)', fontsize=12)
axes[2].set_ylabel('Percentage (%)', fontsize=12)
axes[2].grid(True, linestyle='--', alpha=0.6)
```

```python
# ==========================================
# FINAL FORMATTING & EXPORT
# ==========================================
plt.suptitle('Head-to-Head Performance Across 50 Independent Monte Carlo Simulations', fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust to fit the suptitle

# Save as a high-resolution PNG
plt.savefig('vertical_performance_metrics.png', dpi=300, bbox_inches='tight')
plt.close()
print("Saved: vertical_performance_metrics.png")
```
